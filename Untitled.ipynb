```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from plotnine import ggplot, aes, geom_histogram, geom_bar, labs
from scipy.stats import linregress
from itertools import product
# Dosya okuma
dataset = pd.read_csv("Chennai houseing sale.csv")

# İlk 10 satırı gösterme 
print(dataset.head(10))
```

```python
# Column isimlerini küçük harfe dönüştürme
dataset.columns = dataset.columns.str.lower()

# Verileri küçük harfe dönüştürme
for column in dataset.columns:
    if dataset[column].dtype == 'object':  # Eğer sütun tipi 'object' ise (yani string)
        dataset[column] = dataset[column].str.lower()

# İlk 10 satırı gösterme (küçük harfe dönüşmüş)
print(dataset.head(10))
```

```python
# Boş değer sayısını hesapla
bos_deger_sayisi = dataset.isnull().sum().sum()

# Sonucu ekrana yazdır
print("Boş değer sayısı:", bos_deger_sayisi)
```

```python
# Boş değer içeren sütunları belirleme
columns_with_na = dataset.columns[dataset.isna().any()].tolist()

# Boş değer içeren sayısal sütunları seçme
numeric_columns_with_na = [col for col in columns_with_na if dataset[col].dtype in ['int64', 'float64']]

# Boş değer içeren sütunları ekrana yazdırma
print("Boş değer içeren sütunlar:", columns_with_na)
```

```python
# Boş değerleri medyan ile doldurma
for col in numeric_columns_with_na:
    dataset[col] = dataset[col].fillna(dataset[col].median())

# Eksik değerleri kontrol etme
bos_deger_sayisi = dataset.isnull().sum().sum()

# Boş değerlerin sıfırlandığı sayıyı ekrana yazdırma
print("Boş değerler sıfırlandı sayısı:", bos_deger_sayisi)
```

```python
# Tüm sütun adlarını küçük harfe dönüştürme
dataset.columns = dataset.columns.str.lower()

# area sütunundaki tüm değerleri küçük harfe dönüştürme ve yazım hatalarını düzeltme
dataset['area'] = dataset['area'].str.lower().replace({'velchery': 'velachery', 
                                                       'kknagar': 'kk nagar', 
                                                       'tnagar': 't nagar', 
                                                       'chormpet': 'chrompet', 
                                                       'chrompt': 'chrompet', 
                                                       'chrmpet': 'chrompet', 
                                                       'ana nagar': 'anna nagar', 
                                                       'ann nagar': 'anna nagar',
                                                       'karapakam': 'karapakkam', 
                                                       'adyr': 'adyar'})

# sale_cond sütunundaki tüm değerleri küçük harfe dönüştürme ve yazım hatalarını düzeltme
dataset['sale_cond'] = dataset['sale_cond'].str.lower().replace({'adj land': 'adjland', 
                                                                'normal sale': 'normal sale',
                                                                'partiall': 'partial',
                                                                'ab normal': 'abnormal'})

# park_facil sütunundaki tüm değerleri küçük harfe dönüştürme ve yazım hatalarını düzeltme
dataset['park_facil'] = dataset['park_facil'].str.lower().replace({'noo': 'no'})

# buildtype sütunundaki tüm değerleri küçük harfe dönüştürme ve yazım hatalarını düzeltme
dataset['buildtype'] = dataset['buildtype'].str.lower().replace({'comercial': 'commercial', 
                                                                'others': 'other'})

# utility_avail sütunundaki tüm değerleri küçük harfe dönüştürme ve yazım hatalarını düzeltme
dataset['utility_avail'] = dataset['utility_avail'].str.lower().replace({'all pub': 'allpub',
                                                                       'nosewr': 'nosewa'})

# street sütunundaki tüm değerleri küçük harfe dönüştürme ve yazım hatalarını düzeltme
dataset['street'] = dataset['street'].str.lower().replace({'pavd': 'paved', 
                                                           'noaccess': 'no access'})

# mzzone sütunundaki tüm değerleri küçük harfe dönüştürme
dataset['mzzone'] = dataset['mzzone'].str.lower()

# Doğru Veri Türünü Düzeltme
dataset['n_bedroom'] = dataset['n_bedroom'].astype(int)
dataset['n_bathroom'] = dataset['n_bathroom'].astype(int)

# Object olan Tarih sütununu datetime formatına düzeltme
dataset['date_sale'] = pd.to_datetime(dataset['date_sale'], format="%d-%m-%Y")
dataset['date_build'] = pd.to_datetime(dataset['date_build'], format="%d-%m-%Y")

# property_age sütunu oluşturma
dataset['property_age'] = dataset['date_sale'].dt.year - dataset['date_build'].dt.year

# total_price sütunu oluşturma
dataset['total_price'] = dataset['reg_fee'] + dataset['commis'] + dataset['sales_price']

# Sütunların Yerleşimini Düzenleme
dataset = dataset[['prt_id', 'area', 'sale_cond', 'park_facil',
                   'buildtype', 'utility_avail', 'street', 'mzzone', 
                   'date_build', 'date_sale', 'property_age', 
                   'int_sqft', 'dist_mainroad', 'n_bedroom','n_bathroom', 'n_room', 
                   'qs_rooms', 'qs_bathroom', 'qs_bedroom', 'qs_overall', 
                   'reg_fee', 'commis', 'sales_price', 'total_price']]

print(dataset.head(10))
```

```python
# Boş değerlerin sayısını kontrol etme
bos_deger_sayisi = dataset.isnull().sum().sum()

# Boş değerlerin sıfırlandığı sayıyı ekrana yazdırma
print("Boş değerler sıfırlandı sayısı:", bos_deger_sayisi)
```

```python
# Grafiği oluşturma fonksiyonu
def create_scatterplot(ax, x, y, color):
    sns.scatterplot(x=x, y=y, data=dataset, color=color, ax=ax)
    slope, intercept, r_value, p_value, std_err = linregress(dataset[x], dataset[y])
    ax.plot(dataset[x], slope * dataset[x] + intercept, color='red')

# Grafik düzeni ve dış boşluk ayarları
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
fig.suptitle('Descrete numerical variable VS Total price', fontsize=16)

# Grafikleri çizme
create_scatterplot(axes[0, 0], 'n_bedroom', 'total_price', 'blue')
axes[0, 0].set_xlabel('n_bedroom')
axes[0, 0].set_ylabel('Total Price')

create_scatterplot(axes[0, 1], 'n_bathroom', 'total_price', 'grey')
axes[0, 1].set_xlabel('n_bathroom')
axes[0, 1].set_ylabel('Total Price')

create_scatterplot(axes[0, 2], 'n_room', 'total_price', 'orange')
axes[0, 2].set_xlabel('n_room')
axes[0, 2].set_ylabel('Total Price')

create_scatterplot(axes[1, 0], 'qs_rooms', 'total_price', 'purple')
axes[1, 0].set_xlabel('qs_rooms')
axes[1, 0].set_ylabel('Total Price')

create_scatterplot(axes[1, 1], 'qs_bathroom', 'total_price', 'green')
axes[1, 1].set_xlabel('qs_bathroom')
axes[1, 1].set_ylabel('Total Price')

create_scatterplot(axes[1, 2], 'qs_bedroom', 'total_price', 'black')
axes[1, 2].set_xlabel('qs_bedroom')
axes[1, 2].set_ylabel('Total Price')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
